```python
# ‚Ü™ openFile()
#    Recebe o nome de um ficheiro que contenha informa√ß√£o a ser convertida/lida como pickle.
#       Argumentos:
#       ¬ª filename - Nome do ficheiro a converter;
#       Returns:
#       ¬ª Conte√∫do convertido;
def openFile(filename):
    return pickle.load(open(filename, 'rb'))


# ‚Ü™ limpar()
#    Recebe um array com um ou mais documentos, limpando-os de forma a conter apenas em caracteres
#    alfab√©ticos sem acentos.
#       Argumentos:
#       ¬ª documentos - Array com o(s) documento(s) a limpar;
#       ¬ª binary     - flag, por omiss√£o a 0, que indica necessidade de descodifica√ß√£o bin√°ria
#                      (se o ficheiro estivesse nesse formato);
#       Returns:
#       ¬ª Array com os documentos limpos;
def limpar(documentos, binary=False):
    # Passo 1 ¬ª Converter strings bin√°rias para strings normais
    if binary:
        documentos = [doc.decode('UTF-8') for doc in documentos]
        
    # Passo 2 ¬ª Remo√ß√£o de marcas html de mudan√ßa de linha
    documentos = [doc.replace('<br />', ' ') for doc in documentos]
    
    # Passo 3 ¬ª Eliminar todos os caracteres que n√£o sejam alfab√©ticos
    documentos = [re.sub(r'[^a-zA-Z\u00C0-\u00FF]+' , ' ', doc) for doc in documentos]
    
    return documentos
```

**¬ª Pedido -** Fun√ß√£o que converte uma lista de *strings* (vari√°vel *Docs*) na representa√ß√£o *tf-idf*. Deve retornar uma matriz documento/termo (matriz *X*). A fun√ß√£o dever√° fazer as limpezas necess√°rias aos documentos (*strings*) e usar um modelo *tf-idf* previamente treinado. A matriz ser√°, depois, usada nas fun√ß√µes de classifica√ß√£o bin√°ria e multi-classe.\
**¬ª Implementa√ß√£o -** A fun√ß√£o recebe a lista de *strings*, na referida vari√°vel *Docs*, come√ßando por limp√°-la (chamando a respetiva fun√ß√£o auxiliar). Depois, mediante o tipo de classifica√ß√£o que se pretende obter - algo indicado pelo argumento *tipo_classificacao* - obter√°, ent√£o, de um ficheiro guardado, o modelo *tf-idf* j√° treinado com os melhores par√¢metros encontrados, convertendo, assim, os documentos na matriz *X* pedida, retornando-a.

```python
# ‚Ü™ text2vector()
#    Recebe uma lista de strings, limpando-as e convertendo-as na matriz X, conforme modelo tf-idf
#   previamente treinado.
#       Argumentos:
#       ¬ª Docs               - Lista de string;
#       ¬ª tipo_classificacao - 'b' se bin√°ria, 'm' se multi-classe;
#       ¬ª binary             - flag, por omiss√£o a 0, que indica necessidade de descodifica√ß√£o
#                              bin√°ria (se o ficheiro estivesse nesse formato);
#       Returns:
#       ¬ª Matriz X na representa√ß√£o tf-idf;
def text2vector(Docs, modo = 'b', binary=False):
    Docs  = limpar(Docs, binary)
    tfidf = None
   
    tfidf = openFile('tfidf_binario.p')['tfidf'] if modo == 'b' else openFile('tfidf_multiclasse.p')['tfidf']
        
    if (tfidf):
        X = tfidf.transform(Docs)
        return X
```

- ***binClassify:***  \
**¬ª Pedido -** Fun√ß√£o que classifica a matriz *X* da fun√ß√£o anterior em positivos e negativos. Deve retornar um array, *y*, de 0s e 1s com o resultado da classifica√ß√£o. Pode usar um classificador que achar adequado, mas este j√° deve estar treinado.\
**¬ª Implementa√ß√£o -** O seu funcionamento √© bastante simples - obt√©m, de um ficheiro j√° guardado, o classificador j√° treinado relativo √† classifica√ß√£o bin√°ria, retornando a previs√£o/classifica√ß√£o da matriz *X*.

```python
# ‚Ü™ binClassify()
#    Classifica a matriz X em cr√≠ticas positivas e negativas.
#       Argumentos:
#       ¬ª X                            - Matriz na representa√ß√£o tf-idf;
#       ¬ª nClassificador               - N√∫mero do classificador a usar - por default usa-se o Linear SVC;
#       Returns:
#       ¬ª Classifica√ß√£o bin√°ria;
def binClassify(X, nClassificador = 2):
    classificador = None
    if nClassificador == 1:
        classificador = openFile('modelo_logistic_regression.p')['modelo']
        
    if nClassificador == 2:
        classificador = openFile('modelo_linear_svc_binario.p')['modelo']
        
    if nClassificador == 3:
        classificador = openFile('modelo_knn_binario.p')['modelo']
        
    if nClassificador == 4:
        classificador = openFile('modelo_multinomialNB_binario.p')['modelo']
    return classificador.predict(X)
```

- ***multiClassify:***  \
**¬ª Pedido -** Fun√ß√£o que classifica a matriz *X* devolvida pela fun√ß√£o *text2vector* numa de oito classes (1-4 e 7-10). Deve retornar um array, *y*, de inteiros (1-4 e de 7-10) com o resultado da classifica√ß√£o. Pode usar um classificador que achar adequado, mas este j√° deve estar treinado.\
**¬ª Implementa√ß√£o -** O seu funcionamento √© bastante simples - obt√©m, de um ficheiro j√° guardado, o classificador j√° treinado relativo √† classifica√ß√£o bin√°ria, retornando a previs√£o/classifica√ß√£o da matriz *X*.

```python
# ‚Ü™ multiClassify()
#    Classifica a matriz X nas classes 1-4 e 7-10.
#       Argumentos:
#       ¬ª X                                - Matriz na representa√ß√£o tf-idf;
#       ¬ª nClassificador                   - N√∫mero do classificador a usar;
#       Returns:
#       ¬ª Classifica√ß√£o multiclasse;
def multiClassify(X, nClassificador = 1):
    classificador = None
    if nClassificador == 1:
        classificador = openFile('modelo_logistic_regression_multiclasse.p')['modelo']
        
    if nClassificador == 2:
        classificador = openFile('modelo_linear_svc_multiclasse.p')['modelo']
        
    if nClassificador == 3:
        classificador = openFile('modelo_knn_multiclasse.p')['modelo']
        
    if nClassificador == 4:
        classificador = openFile('modelo_multinomialNB_multiclasse.p')['modelo']
    return classificador.predict(X)
```

### ‚óΩ 1.2 Para que Serve a Vetoriza√ß√£o? - Estudo Preliminar do Tamanho do Vocabul√°rio
**- O que √© *TF-IDF*?**\
*Term frequency-inverse document frequency* √© uma forma que permite medir e quantificar a import√¢ncia que dada palavra tem numa cole√ß√£o de documentos (o *corpus*).\
\
***¬ª TF (Term Frequency):*** Relaciona-se a frequ√™ncia de cada termo em estudo com cada documento do seu geral. A frequ√™ncia pode ser calculada na sua forma mais b√°sica, apenas por contagens, ou tratar, ser mais ajustada/normalizada ao tamanho do documento, ou calculada com m√©todos mais robustos (como de forma logaritmica).\
***¬ª IDF (Inverse Document Frequency):*** Tem em conta o qu√£o comum (ou incomum) uma palavra √© em todo o *corpus*. A import√¢ncia do *IDF* assenta na corre√ß√£o de palavras que aparecem com muita frequ√™ncia na l√≠ngua em que o texto est√° escrito - deve-se maximizar o peso de termos com menos frequ√™ncia.\
\
Quanto mais alto for a pontua√ß√£o/score *TF-IDF*, mais importante/relevante √© esse termo (aproximando-se de 0 quanto menos relevante for).
\
Verifique-se como o ajuste de par√¢metros afetam os dados que temos em m√£o.\

- ***token_pattern:*** Express√£o regular que indica um "padr√£o" que uma palavra deve ter para ser considerada um *token*. Nos exemplos testados, o n√∫mero de '\ws' indica o n√∫mero m√≠nimo de caracteres requisitado.

### ‚óΩ 2.1 Defini√ß√£o, Uso e Metodologia
O primeiro classificador utilizado neste projeto √© o discriminante log√≠stico, que em *sklearn* (apesar de se tratar de um classificador e n√£o se um regressor) tem o nome de *LogisticRegression*.
\
Este classificador consiste numa transforma√ß√£o linear, seguida de uma n√£o-linearidade e √© usado para estimar um conjunto de valores, tendo como dados dispon√≠veis os dados de entrada e as sa√≠das desejadas.\
\
O discriminante log√≠stico √© dado pela fun√ß√£o $y = œÜ(W^T x)$, sendo que o valor mais alto de $y$ traduz a classe escolhida pelo classificador.\
\
√â de salientar ainda o facto deste classificador variar de acordo com os par√¢metros introduzidos, nomeadamente *tol* (Toler√¢ncia), *max_iter* (N√∫mero M√°ximo de Itera√ß√µes), *solver* (M√©todos de Otimiza√ß√£o) e *multi_class* (difere entre classifica√ß√£o bin√°ria e multi-classe).\
\
O discriminante log√≠stico tamb√©m costuma ter um termo de regulariza√ß√£o - em *sklearn* o termo - *penalty* - por omiss√£o costuma ser a regulariza√ß√£o *Ridge* (acrescenta-se a uma fun√ß√£o de custo, outra fun√ß√£o de custo - penalizar pesos com valores de ùëä altos), sendo tamb√©m op√ß√£o a regulariza√ß√£o *Lasso* (esta vai permitir p√¥r v√°rios dos valores dos pesos a 0 - se o vetor de dados (ùë•) tem uma dimens√£o muito elevada, ter v√°rios dos valores a 0 permitir√° escolher quais das dimens√µes s√£o √∫teis e quais n√£o s√£o) e tamb√©m a op√ß√£o da regulariza√ß√£o *Elastic-Net* que √©, basicamente, uma combina√ß√£o das regulariza√ß√µes *Ridge* e *Lasso*. Caso n√£o se deseje ter nenhum termo de regulariza√ß√£o, coloca-se o par√¢metro *penalty = None*. A regulariza√ß√£o deve ser usada quando o modelo tem a capacidade de entrar em sobre aprendizagem.\
\
As vantagens deste classificador passam pelo facto deste ter bons resultados quando o conjunto de dados √© linearmente separ√°vel, ser menos propenso a sobre aprendizagem, mas poder sobre aprender em conjuntos de dados de alta dimens√£o - atrav√©s dos termos de regulariza√ß√£o. O discriminante log√≠stico √© ainda f√°cil de implementar e interpretar e muito eficiente no treino.\
\
J√° as desvantagens passam pelo facto de, no mundo real, os dados raramente serem linearmente separ√°veis e n√£o se verificar linearidade entre a vari√°vel dependente e as vari√°veis independentes. Para al√©m disso, este classificador tem a sua utiliza√ß√£o limitada a problemas cujo conjunto de dados √© composto por n√∫meros discretos, n√£o podendo ser utilizado para estimar dados cont√≠nuos.

### ‚ñ™ 2.2.1 Procura dos Melhores Par√¢metros
O pr√≥ximo passo ser√° encontrar os melhores valores relativos a ***document_frequency, token_pattern, n_grams, C*** e ***Penalty***. Falaremos mais dos *n_gramas* de seguida, mas apresenta-se agora uma fun√ß√£o ***teste_binario*** que:<br><br>
¬ª **Divide os documentos em treino, valida√ß√£o e teste** - Esta divis√£o permite uma averigua√ß√£o menos "brusca" dos resultados.\
¬ª **Recebe uma combina√ß√£o de *document_frequency, token_pattern* e *n_grams*, usando-a para criar o *TDIDF Vectorizer*** a utilizar.\
¬ª Transforma os documentos com essa fun√ß√£o de vetoriza√ß√£o.\
¬ª Procura os pares de *C* e *Penalty* ideiais, de acordo com 4 par√¢metros: **melhor valida√ß√£o cruzada, menor n√∫mero de erros, melhor precis√£o, melhor score.**\
¬ª **Retorna a m√©dia dos scores obtidos com cada um desses pares**, a fim de serem posteriormente comparados de modo a escolher os melhores par√¢metros.\
\
**Valida√ß√£o Cruzada? Precis√£o?**\
A valida√ß√£o cruzada refere uma t√©cnica que procura descobrir qu√£o bem um dado modelo se generaliza para um *dataset* independente. √â uma t√©cnica usada para avaliar este tipo de modelos e estimar o qu√£o adapt√°veis estes ser√£o.\
\
Relativamente √† precis√£o, esta √© outra m√©trica muitas vezes usada pra quantificar a performance de um modelo. Tamb√©m chamado de valor preditivo positivo, √© definido pelo n√∫mero de Verdadeiros Positivos dividido pelo n√∫mero de Positivos Total - ou seja, quantifica a percentagem de previs√µes corretamente identificadas como positivas.

```python
# Fun√ß√£o auxiliar que mostra resultados obtidos da calibra√ß√£o
def avalResultados(criterio, X1t, X1v, X2, y1t, y1v, y2, par):
    print('\n    ‚Ü¨ Crit√©rio:', criterio)
    lr = LogisticRegression(penalty=par[0], solver='liblinear',C=par[1], max_iter=5000, tol=1e-2)
    lr.fit(X1t, y1t)
    
    scoreTrain = lr.score(X1t, y1t)*100
    scoreVal   = lr.score(X1v, y1v)*100
    scoreTest  = lr.score(X2, y2)*100
    print('        ¬ª Acertos (treino):    ', scoreTrain)
    print('        ¬ª Acertos (valida√ß√£o): ', scoreVal)
    print('        ¬ª Acertos (teste) :    ', scoreTest)
    print('        ¬ª Par√¢metros (cv) : penalty='+par[0]+', C=', par[1])
    
    print('        ¬ª Matriz de Confus√£o (teste)\n', confusion_matrix(y2, lr.predict(X2)))
    return scoreVal # Retorna a m√©dia dos scores

def teste_binario(documents, targets, mdf, tp, ngrams):
    print('*. In√≠cio de Novo Teste')
    print('    ‚Ü¨ min_df:', mdf, '  token_pattern:', tp, '  n_grams:', ngrams)
    
    Docs1, Docs2, y1, y2     = sklearn.model_selection.train_test_split(documents, targets, test_size=1/3, stratify=targets)
    Docs1t, Docs1v, y1t, y1v = sklearn.model_selection.train_test_split(Docs1, y1, test_size=1/2, stratify=y1)
    
    print('\n    ‚ü≤ Vetoriza√ß√£o... ')
    tfidf    = TfidfVectorizer(min_df=mdf, token_pattern=tp, ngram_range=ngrams).fit(Docs1t)
    print('    ‚úî Vetoriza√ß√£o conclu√≠da! ')

    tf_info  = {'tfidf' : tfidf}
    
    pickle.dump(tf_info, open('testes_tf_idf/tfidf_test_' + str(mdf) + '_' + str(tp.replace('\\', '').replace('b', '').count('w')) + '_' + str(ngrams) + '.p', 'wb'))
    
    tokens   = tfidf.get_feature_names()
    voc_size = len(tokens)
    print('    ‚Ü¨ Tamanho Vocabul√°rio:', str(voc_size))

    X1, X1t, X1v, X2   = tfidf.transform(Docs1), tfidf.transform(Docs1t), tfidf.transform(Docs1v), tfidf.transform(Docs2)
    
    cList = [0.01, 0.1, 1, 5, 10]
    pList = ['l1', 'l2']
    
    parCVS      = parErros = parPrecision = parScore = None
    menorNErros = 50000
    melhorCVS   = 0
    melhorScore = 0
    melhorDist  = 1
    melhorPonto = None
    print('\n    ‚ü≤ In√≠cio da procura dos melhores par√¢metros... ')
    for p in pList:
        for c in cList:
            lr           = LogisticRegression(penalty=p, solver='liblinear', C=c, max_iter=5000, tol=1e-2)
            
            # Avalia√ß√£o com base no cross_val_score
            cvs          = np.mean(cross_val_score(lr, X1, y1, cv=5))
            
            # Avalia√ß√£o por decision function
            lr.fit(X1t, y1t)
            y1ve         = lr.predict(X1v)
            e            = np.sum(y1v != y1ve)
            
            # Score
            score        = lr.score(X1v, y1v)
            
            # Precis√£o
            prec, rec, _ = precision_recall_curve(y1v, y1ve)
            pontos       = np.array([prec, rec])
            dists        = np.sqrt(np.sum((pontos - np.array([[1.0], [1.0]]))**2, axis=0))
            idx_min      = np.argmin(dists)
            dist         = dists[idx_min]
            ponto_ideal  = pontos[:, idx_min]
                
            if dist <= melhorDist:
                melhorPonto  = ponto_ideal
                parPrecision = (p, c) 
            if score >= melhorScore:
                melhorScore  = score
                parScore     = (p, c) 
            if e <= menorNErros:
                menorNErros  = e
                parErros     = (p, c)
            if cvs >= melhorCVS:
                melhorCVS    = cvs
                parCVS       = (p, c)
    print('    ‚úî Melhores par√¢metros encontrados!')
       
    score1 = avalResultados('Melhor score', X1t, X1v, X2, y1t, y1v, y2, parScore)
    score2 = avalResultados('Menor n√∫mero de erros', X1t, X1v, X2, y1t, y1v, y2, parErros)
    score3 = avalResultados('Melhor cross-validation-score', X1t, X1v, X2, y1t, y1v, y2, parCVS)
    score4 = avalResultados('Melhor valor de precis√£o', X1t, X1v, X2, y1t, y1v, y2, parPrecision)
    
    media  = np.mean(np.array([score1, score2, score3, score4]))
    
    print('\n    ‚Ü¨ Avalia√ß√£o final:', media)
    print('\n\n')
    return (voc_size, media)
```

As c√©lulas seguintes encontram-se comentadas tendo em conta o tempo necess√°rio para todo o processo, mas tinham a fun√ß√£o de chamar o m√©todo criado, testando v√°rias combina√ß√µes de *document_frequency*, *token_pattern* e *n_grams*.<br><br>
*N-grams* s√£o usados para verificar a ocorr√™ncia de palavras numa dada janela. Por exemplo, "Estudo no Instituto Superior de Engenharia de Lisboa". Se *n-2* temos: "estudo no", "no Instituto", "Instituto Superior", etc. Este agrupamento √© importante para simular a obten√ß√£o de contexto de frase - ver a forma como as palavras se situam umas em rela√ß√£o √†s outras e n√£o s√≥ individualmente.<br><br>

### ‚óΩ 4.1 Defini√ß√£o, Uso e Metodologia
O classificador dos k vizinhos mais pr√≥ximos (K-Nearest Neighbors) √© um algoritmo supervisionado que realiza a classifica√ß√£o com base na proximidade de um dado ponto aos seus vizinhos, partindo do pressuposto que pontos semelhantes podem ser encontrados perto uns dos outros.\
\
O KNN √© chamado de algoritmo pregui√ßoso uma vez que n√£o realiza nenhum treino quando lhe s√£o fornecidos os dados de treino. Em vez disso, apenas armazena os dados durante o tempo de treino e n√£o realiza nenhum c√°lculo, n√£o criando nenhum modelo at√© que sejam consultados os dados.\
\
Este classificador possui assim as vantagens de ter uma f√°cil implementa√ß√£o gra√ßas √† sua simplicidade e precis√£o. √â tamb√©m facilmente adapt√°vel, uma vez que √† medida que os dados de treino v√£o sendo adicionados, o algoritmo ajusta-se para considerar quaisquer novos dados, uma vez que estes s√£o armazenados na mem√≥ria. E, por fim, uma outra vantagem deste classificador √© o facto de ter poucos hiperpar√¢metros, precisando apenas do valor de k e de uma m√©trica de dist√¢ncia.\
\
No entanto, o KNN tamb√©m possui algumas desvantagens. Nomeadamente, n√£o funciona bem com elevados conjuntos de dados, uma vez que requer um alto armazenamento de mem√≥ria, nem com dados de dimens√µes elevadas. √â, tamb√©m, necess√°rio determinar o valor de K de acordo com o problema. Para al√©m disso, o processo de estima√ß√£o √© lento se o n√∫mero de vizinhos for elevado.\
\
Com isto em mente, **podem-se esperar, desde j√°, piores resultados do que os obtidos com os classificadores anteriores**.

## üß∑ 2. PCA
<a id='section7-2-1'></a>
### ‚óΩ 2.1 O que trata?
O PCA tem como objetivo projetar dados nas dire√ß√µes de maior vari√¢ncia (as componentes principais), sendo os dados projetados descorrelacionados (a matriz de covari√¢ncia √© diagonal). Pressup√µe-se, assim, que dire√ß√µes onde os dados variam mais cont√™m mais informa√ß√£o.\
\
Consiste ent√£o num m√©todo n√£o supervisionado - n√£o h√° classes - que reduz a dimensionalidade de maneira a preservar o mais poss√≠vel a varia√ß√£o presente nos dados de alta dimens√£o.\
\
**LDA - An√°lise em Discriminantes Lineares**\
O LDA, tamb√©m conhecido por TruncatedSVD, tem como objetivo encontrar uma proje√ß√£o de modo a maximizar a v√¢riancia inter-classe (entre classes) e minimizar vari√¢ncia intra-classe (dentro da mesma classe).\
\
**PCA vs LDA**\
Comparando PCA e LDA destaca-se que no PCA as componentes principais podem n√£o ser as dire√ß√µes mais discriminativas
e que os dados projetados podem ter as mesmas dimens√µes dos dados originais. J√° o LDA possui um melhor poder discriminativo e projeta os dados com c-1 dimens√µes, sendo c o n√∫mero de classes.\
\
Verifiquemos como afeta este tipo de dados.

<a id='section7-2-2'></a>
### ‚óΩ 2.2 Procura de Componentes Ideais - Bin√°rio
√Ä luz dos m√©todos de calibra√ß√£o at√© agora utilizados, testa-se a fun√ß√£o TruncatedSVD com v√°rios n√∫meros de componentes, criando uma fun√ß√£o auxiliar que retorna os resultados da regress√£o log√≠stica ap√≥s tratados com o TruncatedSVD.

```python

def test_components(comps, X1t, X1v, y1t, y1v):
    print('\n\nComponentes ¬ª ', comps)
    svd = TruncatedSVD(n_components=comps, n_iter=5)
    svd.fit(X1t)
    X1tSVD, X1vSVD = svd.transform(X1t), svd.transform(X1v)

    logReg = LogisticRegression(solver='liblinear', penalty='l2', C=3.5, max_iter=5000, tol=1e-2)
    
    logReg.fit(X1tSVD, y1t)
    
    scoret = logReg.score(X1tSVD, y1t)*100
    scorev = logReg.score(X1vSVD, y1v)*100
     
    print('Acertos (treino):    ', scoret)
    print('Acertos (valida√ß√£o): ', scorev)
    
    return (scoret, scorev)

def test_components_multiclasse(comps, X1t, X1v, y1t, y1v):
    print('\n\nComponentes ¬ª ', comps)
    svd = TruncatedSVD(n_components=comps, n_iter=5)
    svd.fit(X1t)
    X1tSVD, X1vSVD = svd.transform(X1t), svd.transform(X1v)

    logReg = LogisticRegression(solver='saga', penalty='l2', C=2.5, max_iter=5000, tol=1e-2, multi_class='multinomial')
    
    logReg.fit(X1tSVD, y1t)
    
    scoret = logReg.score(X1tSVD, y1t)*100
    scorev = logReg.score(X1vSVD, y1v)*100
     
    print('Acertos (treino):    ', scoret)
    print('Acertos (valida√ß√£o): ', scorev)
    
    return (scoret, scorev)
```

### ‚óΩ 3.1 O que trata?
Os m√©todos de agrupamento, ou ***clustering***, s√£o t√©cnicas de aprendizagem n√£o supervisionada onde se procura encontrar representa√ß√µes mais compactas dos dados sem mais nenhuma informa√ß√£o sobre os mesmos. Desta forma pode-se categorizar os dados e descobrir as ‚Äúclasses‚Äù que neles est√£o subjacentes, obtendo assim uma representa√ß√£o de alto n√≠vel dos dados - encontrando-se estes agrupados em nuvens multi-dimensionais de pontos.
